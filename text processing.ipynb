```python
class Preprocess():
    def __init__(self, text, sw=stopwords.words('english'), lower=True, stem=False):
        if not (type(text) == pd.core.series.Series):
            text = pd.Series(text)
        
        self.text = text
        self.sw = sw
        self.lower = lower
        self.stem = stem

    def clean_text(self, text):
        # Remove punctuation
        text = re.sub(r'[^\w\s]', '', text)
        # Convert to lowercase if needed
        if self.lower:
            text = text.lower()
        return text

    def remove_stopwords(self, text):
        return ' '.join([word for word in text.split() if word not in self.sw])

    def preprocess(self):
        cleaned_text = self.text.apply(self.clean_text)
        no_stopwords_text = cleaned_text.apply(self.remove_stopwords)
        return no_stopwords_text
```

```python
# definisi dari label rating
def label_rating(rating):
    if rating in [1, 2]:
        return 'buruk'
    elif rating == 3:
        return 'netral'
    else:  # rating in [4, 5]
        return 'baik'

# pengaplikasian label rating
df['label'] = df['rating'].apply(label_rating)
```

```python
# Penggunaan
file_path = 'reviews_english.csv'
df = pd.read_csv(file_path)

# Menggabungkan kolom menjadi satu
text_columns = ['caption', 'username']  # Adjust based on relevant columns
combined_text = df[text_columns].fillna('').agg(' '.join, axis=1)
```

```python
# Inisialisasi dan lakukan preprocessing tanpa stemming
preprocessor = Preprocess(combined_text, stem=False)
processed_text = preprocessor.preprocess()
```

```python
# Hitung frekuensi kata
all_words = ' '.join(processed_text).split()
word_freq = Counter(all_words)
```

```python
# Buat DataFrame dari frekuensi kata
word_freq_df = pd.DataFrame(word_freq.items(), columns=['word', 'frequency'])
```

```python
# Tampilkan kata yang paling sering muncul
most_common_words = word_freq.most_common
print(most_common_words)
```

```python
from wordcloud import WordCloud
```

```python
# Fungsi untuk mengidentifikasi kata positif dan negatif menggunakan TextBlob
def get_sentiment_words(text):
    positive_words = []
    negative_words = []
    
    for word in text.split():
        analysis = TextBlob(word)
        if analysis.sentiment.polarity > 0:
            positive_words.append(word)
        elif analysis.sentiment.polarity < 0:
            negative_words.append(word)
    
    return ' '.join(positive_words), ' '.join(negative_words)
```

```python
# Dapatkan kata-kata positif dan negatif
all_text = ' '.join(processed_text)
positive_text, negative_text = get_sentiment_words(all_text)
```

```python
# Menampilkan Kata Positif
positive_wordcloud = WordCloud(colormap='Greens', width=1000, height=1000, mode='RGBA', background_color='white').generate(positive_text)

plt.figure(figsize=(10, 10))
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Kata-Kata Positif")
plt.show()
```

```python
netral_wordcloud = WordCloud(colormap='Blues', width=1000, height=1000, mode='RGBA', background_color='white').generate(netral_text)
plt.figure(figsize=(10, 10))
plt.imshow(netral_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Ulasan Netral")
plt.show()
```

```python
# Menampilkan kata negatif
buruk_wordcloud = WordCloud(colormap='Reds', width=1000, height=1000, mode='RGBA', background_color='white').generate(buruk_text)
plt.figure(figsize=(10, 10))
plt.imshow(buruk_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Ulasan Buruk")
plt.show()
```
